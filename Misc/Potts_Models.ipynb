<h1><font color="lightblue">Normalized Discounted Cumulative Gain at K [NDCG@K]</font></h1>

Normalized Discounted Cumulative Gain at k [NDCG@K] is 
With count data, $Pois(\mu_i)$, we sometimes see `overdispersion`, which is when $Var[Y_i] > \mu_i$. We can model this in 4 separate ways. In each case, we give some freedom to the poisson parameter s.t. $Y| \Lambda \sim Poisson(\Lambda)$ where $\mathbb{E}[\Lambda] = \mu$. Since $\Lambda$ is no longer deterministic, we need to pick a distribution for

1. `Mutual Independence: Pr(A=i,B=j,C=k) = Pr(A=i)*Pr(B=j)*Pr(C=k)`
* ALL ARE INDEPENDENT


<div align="center">

$$
log\mu_{ijk} = \beta_0 + \beta_i^A + \beta_j^B + \beta_k^C \implies \pi_{ijk} = \pi_{i++}\pi_{+j+}\pi_{++k}
$$

</div>

```python
import numpy as np

def dcg_at_k(relevance, k):
    """Compute DCG@k"""
    relevance = np.array(relevance)[:k]
    discounts = np.log2(np.arange(2, len(relevance) + 2))
    return np.sum(relevance / discounts)


def ndcg_at_k(y_true, y_pred, k):
    """
    y_true: true relevance scores (list)
    y_pred: predicted scores (same length)
    k: cutoff
    """
    # Sort by predicted scores (descending)
    order = np.argsort(y_pred)[::-1]
    y_true_sorted = np.array(y_true)[order]

    # Compute DCG
    dcg = dcg_at_k(y_true_sorted, k)

    # Compute IDCG (ideal ranking)
    ideal_order = np.argsort(y_true)[::-1]
    y_true_ideal = np.array(y_true)[ideal_order]
    idcg = dcg_at_k(y_true_ideal, k)

    if idcg == 0:
        return 0.0

    return dcg / idcg


# 🔬 Example usage
y_true = [3, 2, 3, 0, 1]
y_pred = [0.2, 0.1, 0.4, 0.05, 0.3]

k = 5

print("NDCG@{}: {:.4f}".format(k, ndcg_at_k(y_true, y_pred, k)))
```
