# <font color='lightblue'> Operators </font>

> Summation Operator

The summation operator gives us a compact way to represent long sums over an collection of variables and is described as follows:

<div align="center">

$$
\sum_{i=1}^n Y_i= Y_1 + Y_2 + ... + Y_n
$$

</div>

These are the important properties to remember: 
1. <font color='red'>Constants multiply</font>: $\sum_{i=1}^nk = k + k + ... + k = nk$
2. <font color='red'>Distributive property</font>: $\sum_{i=1}^n (Y_i +Z_i) = \sum_{i=1}^n Y_i + \sum_{i=1}^n Z_i$
3. <font color='red'>Linearity property</font>: $\sum_{i=1}^n (a + cY_i) = na + c \sum_{i=1}^n (Y_i)$
4. <font color='red'>Double summation</font>: $\sum_{i=1}^n \sum_{j=1}^m Y_{i,j} = \sum_{i=1}^n (Y_{i,1} + Y_{i,2} + ... + Y_{i,m} ) = Y_{1,1} + ... + Y_{1,m} + Y_{2,1} + ... + Y_{2,m} + ... + Y_{n,m}$
5. <font color='red'>Commutative property</font>: $\sum_{i=1}^n \sum_{j=1}^m Y_{i,j} = \sum_{j=1}^m \sum_{i=1}^n Y_{i,j} $


<font color='purple'>Generally, the subscript tells what index to iterate over, and letters without a subscript are typically constants</font>

$$

> Expectation Operator

By definition, the expecation of a random variable is related to the probability density function (PDF), $\mathcal{f}(Y)$ as follows: 


<div align="center">

$$
\mathbb{E} [Y] = \int_{-\infty}^\infty Y \mathcal{f}(Y) dY
$$

</div>

These are the important properties to remember: 
1. <font color='red'>Constants carry</font>: $\mathbb{E} [a] = a$
2. <font color='red'>Factorable</font>: $\mathbb{E} [aY] = a \mathbb{E} [Y]$
3. <font color='red'>Linearity property</font>: $\mathbb{E} [X+Y] = \mathbb{E} [X] + \mathbb{E} [Y]$

<font color='purple'> Note, for a discrete RV, this becomes the probability mass function (PMF) instead, and we can swap integrals for summation operator</font>

# <font color='lightblue'> Random Variables </font>

> Discrete vs Continuous

A random variable is a variable, typically written using capital letters, where the value is unknown prior to being observed. For example, if we define $X$ to be the outcome from flipping a coin, then $X \in \{heads, tails\}$ is a random variable since we do not know which one it will be until we actually flip said coin. These RVs can be either `discrete` (countable) or `continuous` (range).
