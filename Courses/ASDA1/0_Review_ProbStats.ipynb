# <font color='lightblue'> Operators </font>

> Summation Operator

The summation operator gives us a compact way to represent long sums over an collection of variables and is described as follows:

<div align="center">

$$
\sum_{i=1}^n Y_i= Y_1 + Y_2 + ... + Y_n
$$

</div>

These are the important properties to remember: 
1. <font color='red'>Constants multiply</font>: $\displaystyle \sum_{i=1}^nk = k + k + ... + k = nk$
2. <font color='red'>Distributive property</font>: $\displaystyle \sum_{i=1}^n (Y_i +Z_i) = \sum_{i=1}^n Y_i + \sum_{i=1}^n Z_i$
3. <font color='red'>Linearity property</font>: $\displaystyle \sum_{i=1}^n (a + cY_i) = na + c \sum_{i=1}^n (Y_i)$
4. <font color='red'>Double summation</font>: $\displaystyle \sum_{i=1}^n \sum_{j=1}^m Y_{i,j} = \sum_{i=1}^n (Y_{i,1} + Y_{i,2} + ... + Y_{i,m} ) = Y_{1,1} + ... + Y_{1,m} + Y_{2,1} + ... + Y_{2,m} + ... + Y_{n,m}$
5. <font color='red'>Commutative property</font>: $\displaystyle \sum_{i=1}^n \sum_{j=1}^m Y_{i,j} = \sum_{j=1}^m \sum_{i=1}^n Y_{i,j} $


<font color='purple'>Generally, the subscript tells what index to iterate over, and letters without a subscript are typically constants</font>


--- 


> Product Operators

This operator gives us a compact way to represent numerous multiplications over a collection of variables and is describes as follows: 


<div align="center">

$$
\prod_{i=1}^n Y_i= Y_1 \cdot Y_2 \cdot ... \cdot Y_n
$$

</div>

--- 

> Expectation Operator

The expecatation operator is a way to measure *centrality of data*. By definition, the expecation of a random variable is related to the probability density function (PDF), $\mathcal{f}(Y)$, as follows: 


<div align="center">

$$
\mathbb{E} [Y] = \int_{-\infty}^\infty Y \mathcal{f}(Y) dY
$$

</div>

These are the important properties to remember: 
1. <font color='red'>"Identity of Constants"</font>: $\mathbb{E} [a] = a$
2. <font color='red'>Constants Factor</font>: $\mathbb{E} [aY] = a \mathbb{E} [Y]$
3. <font color='red'>Linearity property</font>: $\mathbb{E} [X+Y] = \mathbb{E} [X] + \mathbb{E} [Y]$
4. <font color='red'>Iterative Expectations</font>: $\mathbb{E} [X] = \mathbb{E} [\mathbb{E} [X|Y]]$

<font color='purple'> Note, for a discrete RV, this becomes the probability mass function (PMF) instead, and we can swap integrals for summation operator</font>

--- 


> Covariance Operator

The covariance operator is a way to measure *linear relationships of data* by measuring how random variables change with respect to one another and is describes as follows: 


<div align="center">

$$
\sigma(X,Y) = \mathbb{E}[ (X - \mathbb{E}[X]) (Y - \mathbb{E}[Y])]
$$

</div>


`Equivalent Expression`: $\mathbb{E}[ (X - \mathbb{E}[X]) (Y - \mathbb{E}[Y])] \rightarrow$ <font color='green'> Foil, Linearity property, Factorable property </font> $ \rightarrow \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y]$

These are the important properties to remember: 
1. <font color='red'>Center Invariant</font>: $\sigma (X,Y+C) = \sigma (X,Y)$
2. <font color='red'>Linear Factorable</font>: $\sigma (X,cY) = c \cdot \sigma (X,Y)$

<font color='purple'> Note From here on out, the covariance will be referred to using subscripts of the two random variables being measured against: </font> $\sigma_{X,Y}$

--- 

> Variance Operator

The variance operator is a way to measure *spread of data* and can be calculated as follows: 


<div align="center">

$$
\sigma(X,X) = \mathbb{E}[ (X - \mathbb{E}[X]) (X - \mathbb{E}[X])] = \mathbb{E}[ (X - \mathbb{E}[X])^2] = \sigma^2(X)
$$

</div>

`Equivalent Expression`: $(X - \mathbb{E}[X]) (X - \mathbb{E}[X]) \rightarrow$ <font color='green'> Foil, Linearity property, Factorable property </font> $ \rightarrow \mathbb{E}[X^2] - \mathbb{E}[X]^2$

These are the important properties to remember: 
1. <font color='red'>Center Invariant</font>: $\sigma^2 (X+C) = \sigma^2 (X)$
2. <font color='red'>Quadratic Factorable</font>: $\sigma^2 (cX) = c^2 \sigma (X)$
3. <font color='red'>Sum of Variances Identity I</font>: $\sigma^2 (X+Y) = \sigma^2 (X) + \sigma^2 (Y) + 2 \sigma (X,Y)$
4. <font color='red'>Sum of Variances Identity II</font>: $\sigma^2 (X-Y) = \sigma^2 (X) + \sigma^2 (Y) - 2 \sigma (X,Y)$

<font color='purple'> Note From here on out, the variance will not have any subscripts since it is measured with respect to one RV and itself: </font> $\sigma_X^2$

# <font color='lightblue'> Probability Basics</font>

> Random Variables

A random variable, typically written using capital letters, is a variable where the value is unknown prior to being observed. Once the value is *observed* or *fixed* to a specific value, we use a lower case letter to represent it. For example, if we define $X$ to be the outcome from flipping a coin, then $X \in \{heads, tails\}$ is a RV since we do not know a priori (before the flip) and $x_1 = heads$ means on the $1^{st}$ flip, we observed heads. RVs can be `discrete` (countable) or `continuous` (range).


---


> Total Probability

If we consider all possible outcomes (non-overlapping), then one of those cases must occur, so their probabilities add to 100%

<div align="center">

$$
\int_{-\infty}^{\infty} P(A) = 1
$$

</div>

---

> Mutually Exclusive

Super simple, but super important: <font color='red'> A and B can never happen together </font>

<div align="center">

$$
P(A \cap B)=0
$$

</div>

---

> Independence

Super simple, but super important: <font color='red'> knowing whether B happens does not change the probability of A </font>

<div align="center">

$$
P(A|B) = P(A)
$$

</div>

---


> Marginal Probability Distribution

A marginal probability is the usual probability that we think of when we want to measure the "chances of an event occuring" while ignoring all the other variables, $P(A)$. It is found by summing the probability of $A$, over all possible $B's$ ("summing out B")

<div align="center">

$$
P(A) = \int f(A,B)dB
$$

</div>

---

> Joint Probability Distribution

This represents the probability of observing two events happening simultaneously. This has two notations as written below (but we will use the later)

<div align="center">

$$
P(A \cap B) = P(A,B) 
$$

</div>

--- 


> Conditional Probability Distribution

This is the probability of event A, given that we have already observed event B. To get this, we need to narrow the sample space to only outcomes where B has happened.


<div align="center">

$$
P(A | B) = \frac{P(A \cap B)}{P(B)}
$$

</div>

---


> Addition Theorem

We now define the probability of A or B or both, $P(A \cup B)$.


<div align="center">

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

</div>

<font color='purple'>This is the general case! Note what happens if A and B are mutually exclusive</font>

--- 


> Multiplication Theorem 


<div align="center">

$$
P(A \cap B) = P(A) P(B|A)  = P(B) P(A|B)
$$

</div>


<font color='purple'>This is the general case! Note what happens if A and B are independent</font>


---


> Complement 

This is the probability of an event NOT happening.

<div align="center">

$$
P(\overline{A}) = 1 - P(A)
$$

$$
P(\overline{A \cup B}) = P(\bar{A} \cap \bar{B})
$$

</div>


--- 

Probability Table 

<div align="center">

|           |      $B$       |   $\bar{B}$            |    Total   |
|-----------|----------------|------------------------|------------|
|    $A$    |     $P(A,B)$   |  $P(A,\bar{B})$        |   $P(A)$   | 
| $\bar{A}$ | $P(\bar{A},B)$ |  $P(\bar{A},\bar{B})$  |$P(\bar{A})$|  
|   Total   |     $P(B)$     |  $P(\bar{B})$          |      1     |  

</div>

These are the important properties to remember: 
1. <font color='red'> Conditionals </font>: $P(A|B)$ is not explicitly in the table, instead they are calculated by renormalizing by a row (to get $P(B|A)$) or column (to get $P(A|B)$).
2. <font color='red'> Marginals </font>: From `joints` we can get marginals (definition), but from `conditionals` we can NOT go directly to marginals (table)
3. <font color='red'> Total Probability</font>: The bottom right must sum to 1 across the entire probability space



# <font color='lightblue'> Descriptive Statistics </font>

> Degrees of Freedom

Degrees of freedom are the number of values that are still free to vary after building a model from data. If we have N data points and p parameters then:

<div align="center">

$$
DoF = N - p
$$

</div>

DoFs represent how many independent pieces of information are left after estimations (since each estimate contains information about the data). For example, let's say 5 numbers are known to have a mean of 10. Our $DoF = 5 - 1 = 4$, which means we can only vary 4 numbers and the last one is fixed.

---

> Sample Mean

This represents the probability of observing two events happening simultaneously. This has two notations as written below (but we will use the later)

<div align="center">

$$
\bar{Y} = \frac{1}{n} \sum_{i=1}^n Y_i
$$

</div>

These are the important properties to remember: 
1. <font color='red'>Sample mean estimates population mean</font>: $\displaystyle \mathbb{E} [\bar{Y}] = \mathbb{E} [\frac{1}{n} \sum_{i=1}^n Y_i] = \frac{1}{n} \sum_{i=1}^n \mathbb{E} [ Y_i] = \frac{1}{n} \sum_{i=1}^n  \mu = \frac{1}{n} n \mu = \mu $
2. <font color='red'>Sample mean error is proportional to n</font>: $\displaystyle Var[\bar{Y}] = Var[\frac{1}{n} \sum_{i=1}^n Y_i] = \frac{1}{n^2} \sum_{i=1}^n Var[Y_i] = \frac{1}{n^2} \sum_{i=1}^n \sigma^2 = \frac{1}{n^2} n \sigma^2 = \frac{\sigma^2}{n}$

<font color='purple'> Note: In 2, we assume that each observation is independently perturbed from the mean, so cov(X,Y)=0</font>

> Sample Variance

This represents the probability of observing two events happening simultaneously. This has two notations as written below (but we will use the later)

<div align="center">

$$
\hat{\sigma}^2 = S^2 = \frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})^2
$$

</div>

These are the important properties to remember: 
1. <font color='red'>Unbiased Estimator</font>: Since the equation requires the mean, we have LOST a degree of freedom and must account for the in the denominator
2. <font color='red'>Expectation returns varaince</font>: $\displaystyle \mathbb{E}[S^2] = \frac{1}{n-1} \mathbb{E}[\sum_{i=1}^n (X_i - \bar{X})^2] = \frac{1}{n-1} \sum_{i=1}^n \mathbb{E}[(X_i - \bar{X})^2] = \frac{1}{n-1} \sum_{i=1}^n \mathbb{E}[(X_i - \bar{X})^2]$

$Var[\frac{1}{n} \sum_{i=1}^n Y_i] = \frac{1}{n^2} \sum_{i=1}^n Var[Y_i] = \frac{1}{n^2} \sum_{i=1}^n \sigma^2 = \frac{1}{n^2} n \sigma^2 = \frac{\sigma^2}{n}$



---


> Summary Table

<div align="center">

| | Population (N) | Sample (n) |
|---|----------|----------|
|Mean| $\displaystyle \mu = \frac{1}{N} \sum_{i=1}^n X_i$ | $\displaystyle \bar{X} = \frac{1}{n} \sum_{i=1}^n X_i$ |
|Variance| $\displaystyle \sigma^2 = \frac{1}{N} \sum_{i=1}^n (X_i - \mu)^2$ | $\displaystyle  S^2 = \frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})^2$ |
|Std Error | $\displaystyle \sigma = \sqrt{\frac{1}{N} \sum_{i=1}^n (X_i - \mu)^2}$ | $\displaystyle SE = \sqrt{\frac{1}{n-1} \sum_{i=1}^n (X_i - \bar{X})^2}$ |

</div>

# <font color='lightblue'> Special Distributions </font>

> Normal Distribution

It runs everything, here is the PDF for a normally distributed random variable

<div align="center">

$$
f(X| \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} exp \{ \frac{-(X - \mu)^2}{2\sigma^2} \}
$$

</div>

with support from $-\infty, \infty$

<div align="center">

$$
X \sim N(\mu_1, \sigma_1) ; Y \sim N(\mu_2, \sigma_2)
$$

$$
aX + bY \sim N(a\mu_1 + b\mu_2, a^2\sigma_1 + b^2\sigma_2)
$$


</div>


---

> Standard Normal Distribution

It runs everything, here is the PDF for a normally distributed random variable

<div align="center">

$$
z \sim N(\mu=0,\sigma^2=1)
$$

$$
z = \frac{X - \mu}{\sigma}
$$

</div>

---

> Chi-Squared Distribution

Where $\nu$ is the degrees of freedom

<div align="center">

$$
\Chi_{\nu}^2 = \sum_{i=1}^{\nu} z_i^2
$$

</div>

These are the important properties to remember: 
1. <font color='red'>DoF Expectation</font>: $\mathbb{E} [\Chi_{\nu}^2] = \nu$


---

> t Distribution

This represents the probability of observing two events happening simultaneously. This has two notations as written below (but we will use the later)

<div align="center">

$$
t_\nu = \frac{z}{[\frac{\Chi_{\nu}^2}{\nu}]^{1/2}}
$$

</div>

These are the important properties to remember: 
1. <font color='red'>"Identity of Constants"</font>: $\mathbb{E} [a] = a$
2. <font color='red'>Constants Factor</font>: $\mathbb{E} [aY] = a \mathbb{E} [Y]$
3. <font color='red'>Linearity property</font>: $\mathbb{E} [X+Y] = \mathbb{E} [X] + \mathbb{E} [Y]$
4. <font color='red'>Iterative Expectations</font>: $\mathbb{E} [X] = \mathbb{E} [\mathbb{E} [X|Y]]$


---

> F Distribution

Used to compare variances between regression models 

<div align="center">

$$
U \sim \Chi_k^2 ; V \sim \Chi_m^2
$$

$$
F_{k,m} \sim \frac{\frac{U}{k}}{\frac{V}{m}} = \frac{Um}{Vk}
$$

</div>
