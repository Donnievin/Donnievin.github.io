# <font color='lightblue'> Probability</font>

> Goal

* *Statistical Thermodynamics*: a set of tools for modeling molecular forces and behaviors, and a language for interpreting results.

--- 

> Probability

$$
\mathbb{P}_A = \frac{n_A}{N} = \frac{\text{number of outcomes that falls into category A}}{\text{total number of outcomes}}
$$

---

> Deterministic:

Only 1 outcome is possible

$$
\mathbb{P}_A =1
$$

---

> Mutually Exclusive

A and B can <font color='red'> never happen together </font>

<div align="center">

$$
\mathbb{P}_{A \cap B}=0
$$

</div>

---


> Collectively Exhaustive

A set, S, that includes <font color='red'>all possible outcomes</font>

<div align="center">

$$
S = \cup_{i=1}^n A_i
$$

</div>

---

> Multiplicity

The total number of ways different outcomes can occur 

$$
W_{tot} = \prod_{i=events} n_i
$$ 

---

> Independent 

Outcomes are unrelated (not correlated). Knowing if B happened <font color='red'>does not change the probabiity of A</font>

$$
\mathbb{P}_{A | B} = \mathbb{P}_A
$$

<font color='purple'>Note: Independence implies no correlation, but no correlation does not imply independence (e.g. y = x^2)</font>

---

> Joint Probabiliity

Probability of observing A and B <font color='red'>simulatneously</font>

$$
\mathbb{P}_{A,B}
$$


---

> Conditional Probabiliity

Probability of observing A, given that B has already happened. This <font color='red'>restricts the probability space</font>

$$
\mathbb{P}_{A|B}
$$


---


> Addition Rule

Probability of observing either A <font color='red'>OR</font> B, without double counting the overlap

$$
\mathbb{P}_{A \cup B} = \mathbb{P}_A + \mathbb{P}_B - \mathbb{P}_{A \cap B}
$$

<font color='purple'>This is the general case! Note what happens if A and B are mutually exclusive</font>

--- 

> Multiplication Rule (Bayes' Rule)

Probability of observing either A <font color='red'>AND</font> B

$$
\mathbb{P}_{A \cap B} = \mathbb{P}_A \mathbb{P}_{B|A} = \mathbb{P}_B \mathbb{P}_{A|B}
$$

<font color='purple'>This is the general case! Note what happens if A and B are independent</font>

---


> Composite events: 

A set of 2+ independent elementary events (e.g. coin flip, dice roll) that are collected together. [Independent events but of the or type]


* Ex 1: 1 on first roll or 4 on second roll = 11/36
* Ex 2: Tails on the first flip or heads on the third flip = 6/8
* Ex 3: If I have 23 people in a room, what is the probability that 2 will share a birthday = 50% [<font color='green'>Birthday paradox</font>]

<font color='purple'>Can also be rephrased as *multiplication of failures*</font> [see Ex 1.8 for a classic problem]


---

> Degree of Correlation

Indicates the extent to which <font color='red'> A influences B </font>

$$
g = \frac{\mathbb{P}_{B|A}}{\mathbb{P}_B} = \frac{\mathbb{P}_{AB}}{\mathbb{P}_A \mathbb{P}_B}
$$

---

> Complement

The probability that an even does NOT occur

$$
\mathbb{P}_{A'} = 1 - \mathbb{P}_A
$$

---

> Complement Rule

$$
\mathbb{\bar{P}}_{A \cup B} = \mathbb{\bar{P}}_{A \cap B} 
$$

---

> Adding joint probabilities


$$
\mathbb{P}_B = \mathbb{P}_{BA} + \mathbb{P}_{BA'} = \mathbb{P}_{B|A} \mathbb{P}_A + \mathbb{P}_{B|A'} \mathbb{P}_{A'}
$$

# <font color='lightblue'> Combinatorics</font>

> Goal

Combinatorics counts the number of ways events can be configured. It focuses on <font color='red'>composition rather than sequence</font> of events.

* Ex 1: What is the probability of HTHT? $\implies$ specific <font color='red'>sequence</font> question
* Ex 2: What is the probability of 2 Hs and 2 Ts? $\implies$ arbitrary <font color='red'>composition</font> question

--- 

> Permutations

For a collection of N objects in t distinguishable categories (or groups), the number of permutations is:

$$
W = \frac{N!}{n_1! n_2! ... n_t!} = \frac{N!}{\prod_{i} n_i}
$$

--- 

> Binomial Coefficient

If there are only 2 categories (success and failure), we can count the number of successes, n, out of N trials as: 

$$
W(n,N) = \binom{N}{n} = \frac{N!}{n! (N-n)!}
$$

<font color='purple'>Note: Read as "N chose n"</font>

---

> Probability Density Function (PDF)

Used for *continuous* probability functions [mass is for *discrete*]. 

$$
\int_a^b p(x)dx = 1
$$

<font color='purple'>Note: p(x)dx = Probability </font>

---

> Normalization
 
 When PDFs do not sum to 1, we can integrate over the entire range, then divide by this value

$$
p(x) = \frac{g(x)}{g_0} = \frac{g(x)}{\int_a^b g(x)dx}
$$

---

> Binomial Distribution

Gives us the probability of observing n successes, with probability $\mathbb{P}$, in N trials of a binary outcome:

$$
\text{Bin}(n,N) = \binom{N}{n} \mathbb{P}^n (1-\mathbb{P})^{N-n}
$$

<font color='purple'>Note: Each trial (w/ 2 possible outcomes) is called a Bernoulli trial</font>


---

> Multinomial Distribution 

A generalization of the binomial distribution

$$
\text{Multi}(n_1, n_2, ..., n_t, N) = \prod_i \text{Bin}(n_i,N) = \frac{N!}{n_1! n_2! ... n_t!} \mathbb{P}_1^{n_1} \mathbb{P}_2^{n_2} ... \mathbb{P}_t^{n_t}
$$



<font color='purple'>Probabilities are</font> $\mathbb{P}$<font color='purple'>, PDFs are</font> $p(x)$

# <font color='lightblue'> Statistics </font>

> Goal

PDFs contain all information about a system, we just need to extract it. However, in practice we rarely ever have the entire PDF.

---

> Moments

The nth moment of a PDF is defined as:

$$
\langle x^n \rangle = \int_a^b x^n p(x)dx = \frac{\int_a^b x^n g(x)dx}{\int_a^b g(x)dx}
$$

<font color='purple'>Note: The second equation is just a normalization</font>

---

> Mean, Average, or Expected Value

This is the <font color='red'>1st moment</font> of the PDF. This tells us where our data is centered

$$
\langle x \rangle = \int_a^b x p(x)dx
$$


Properties: 
* Property 1: <font color='red'>"Identity"</font> $\langle a \rangle = a $
* Property 2: <font color='red'>"Factorability"</font> $\langle ax \rangle = a \langle x \rangle $
* Property 3: <font color='red'>"Linearity"</font> $\langle ax + by \rangle = a\langle x \rangle + b\langle y \rangle $

--- 

> Variance (Centered)

This is the <font color='red'>2nd moment</font> of the PDF. This tells us how our data is spread

$$
\sigma^2 = \langle [x- \langle x \rangle]^2 \rangle = \langle x^2 \rangle - \langle x \rangle^2
$$

Properties: 
* Property 1: <font color='red'>"Center Invariance"</font> $\sigma^2(x + a) = \sigma^2(x)$
* Property 2: <font color='red'>"Factorability"</font> $\sigma^2(ax) = a^2\sigma^2(x)$
* Property 1: <font color='red'>"Sum of Variance Identity"</font> $\sigma^2(x + y) = \sigma^2(x) + \sigma^2(x) - 2 \sigma(x,y)$ [covariance]

<font color='purple'>Note: This will be important for heat capacities, random walks, diffusion, polymer conformations, and more </font>

--- 

> Skewness (Centered)

This is the 3rd moment of the PDF. This tells us about asymmetries.

$$
\sigma^3 = \langle [x- \langle x \rangle]^3 \rangle
$$

---

> Kurtosis (Centered)

This is the 4th moment of the PDF. This tells us about the thickness of the tails.

$$
\sigma^4 = \langle [x- \langle x \rangle]^4 \rangle
$$

```python
# Doing all of these by hand 
n = np.arange(-5,5,0.1)

## Bernoulli [g(n) = p^n * (1-p)^(N-n)]
## Binomial [p(n) = (N choose n) * p^n * (1-p)^(N-n)]
## p is the probability of success
g_bernoulli_0p2 = (0.2**n) * ((1-0.2)**(1-n)) 
g_bernoulli_0p4 = (0.4**n) * ((1-0.4)**(1-n)) 
g_bernoulli_0p6 = (0.6**n) * ((1-0.6)**(1-n)) 
g_bernoulli_0p85 = (0.85**n) * ((1-0.85)**(1-n)) 

## Poisson [p(n) = (lambda^n * e^-lambda) / n!] [Taylor expansion of e^x]
## Lambda is the average number of events in a given time interval
p_poisson_0p2 = (0.2**n * np.exp(-0.2)) / factorial(n)
p_poisson_0p4 = (0.4**n * np.exp(-0.4)) / factorial(n)
p_poisson_0p6 = (0.6**n * np.exp(-0.6)) / factorial(n)
p_poisson_0p85 = (0.85**n * np.exp(-0.85)) / factorial(n)

## Gaussian [p(n) = (1/(sigma * sqrt(2pi))) * e^(-(n-mu)^2 / (2*sigma^2))]
## Here, we are using mu=0 and varying sigma
p_gaussian_0p2 = (1/(0.2 * np.sqrt(2*np.pi))) * np.exp(-((n-0)**2) / (2*(0.2**2)))
p_gaussian_0p4 = (1/(0.4 * np.sqrt(2*np.pi))) * np.exp(-((n-0)**2) / (2*(0.4**2)))
p_gaussian_0p6 = (1/(0.6 * np.sqrt(2*np.pi))) * np.exp(-((n-0)**2) / (2*(0.6**2)))
p_gaussian_0p85 = (1/(0.85 * np.sqrt(2*np.pi))) * np.exp(-((n-0)**2) / (2*(0.85**2))) 

## Gaussian [p(n) = (1/(sigma * sqrt(2pi))) * e^(-(n-mu)^2 / (2*sigma^2))]
## Here, we are using sigma=1 and varying mu
p_gaussian_m_0p2 = (1/(np.sqrt(2*np.pi))) * np.exp(-((n-0.2)**2) / (2))
p_gaussian_m_0p4 = (1/(np.sqrt(2*np.pi))) * np.exp(-((n-0.4)**2) / (2))
p_gaussian_m_0p6 = (1/(np.sqrt(2*np.pi))) * np.exp(-((n-0.6)**2) / (2))
p_gaussian_m_0p8 = (1/(np.sqrt(2*np.pi))) * np.exp(-((n-0.8)**2) / (2))


## Exponential (Boltzmann) [p(n) = beta * e^(-beta*n)]
## Beta is the inverse temperature (1/kT)
p_exponential_0p2 = 0.2 * np.exp(-0.2*n)
p_exponential_0p4 = 0.4 * np.exp(-0.4*n)
p_exponential_0p6 = 0.6 * np.exp(-0.6*n)
p_exponential_0p85 = 0.85 * np.exp(-0.85*n)

## Power Law [p(n) = 1 / x^a]
p_powerlaw_0p2 = 1 / n**(0.2)
p_powerlaw_0p4 = 1 / n**(0.4)
p_powerlaw_0p6 = 1 / n**(0.6)
p_powerlaw_0p85 = 1 / n**(0.85)

## Lorentzian (centered at 0) [p(n) = 1 / (pi * (x^2 + a^2)]
## 2a = FWHM (full width at half maximum)
p_lorentzian_0p2 = 1 / (np.pi * (0.2**2 + n**2))
p_lorentzian_0p4 = 1 / (np.pi * (0.4**2 + n**2))
p_lorentzian_0p6 = 1 / (np.pi * (0.6**2 + n**2))
p_lorentzian_0p85 = 1 / (np.pi * (0.85**2 + n**2))

## Flat [p(n) = 1 / a]
p_flat_0p2 = [1 / 0.2 for i in n]
p_flat_0p4 = [1 / 0.4 for i in n]
p_flat_0p6 = [1 / 0.6 for i in n]
p_flat_0p85 = [1 / 0.85 for i in n]

fig, ax = plt.subplots(2,4,figsize=(12,7))

ax[0, 0].plot(n, g_bernoulli_0p2, color='orange', label='p=0.2')
ax[0, 0].plot(n, g_bernoulli_0p4, color='yellow', label='p=0.4')
ax[0, 0].plot(n, g_bernoulli_0p6, color='green', label='p=0.6')
ax[0, 0].plot(n, g_bernoulli_0p85, color='blue', label='p=0.85')
ax[0, 0].set_title("Bernoulli")
ax[0, 0].legend()


ax[0, 1].plot(n, p_poisson_0p2, color='orange', label='λ=0.2')
ax[0, 1].plot(n, p_poisson_0p4, color='yellow', label='λ=0.4')
ax[0, 1].plot(n, p_poisson_0p6, color='green', label='λ=0.6')
ax[0, 1].plot(n, p_poisson_0p85, color='blue', label='λ=0.85')
ax[0, 1].set_title("Poisson")
ax[0, 1].legend()

ax[0, 2].plot(n, p_gaussian_0p2, color='orange', label='σ=0.2')
ax[0, 2].plot(n, p_gaussian_0p4, color='yellow', label='σ=0.4')
ax[0, 2].plot(n, p_gaussian_0p6, color='green', label='σ=0.6')
ax[0, 2].plot(n, p_gaussian_0p85, color='blue', label='σ=0.85')
ax[0, 2].set_title("Gaussian (variance)")
ax[0, 2].legend()

ax[0, 3].plot(n, p_gaussian_m_0p2, color='orange', label='μ=0.2')
ax[0, 3].plot(n, p_gaussian_m_0p4, color='yellow', label='μ=0.4')
ax[0, 3].plot(n, p_gaussian_m_0p6, color='green', label='μ=0.6')
ax[0, 3].plot(n, p_gaussian_m_0p8, color='blue', label='μ=0.8')
ax[0, 3].set_title("Gaussian (mean)")
ax[0, 3].legend()

ax[1, 0].plot(n, p_exponential_0p2, color='orange', label='β=0.2')
ax[1, 0].plot(n, p_exponential_0p4, color='yellow', label='β=0.4')
ax[1, 0].plot(n, p_exponential_0p6, color='green', label='β=0.6')
ax[1, 0].plot(n, p_exponential_0p85, color='blue', label='β=0.85')
ax[1, 0].set_title("Exponential (Boltzmann)")
ax[1, 0].legend()

ax[1, 1].plot(n, p_powerlaw_0p2, color='orange', label='α=0.2')
ax[1, 1].plot(n, p_powerlaw_0p4, color='yellow', label='α=0.4')
ax[1, 1].plot(n, p_powerlaw_0p6, color='green', label='α=0.6')
ax[1, 1].plot(n, p_powerlaw_0p85, color='blue', label='α=0.85')
ax[1,1].set_xscale('log')
ax[1, 1].set_title("Power Law (Log X-axis)")
ax[1, 1].legend()  

ax[1, 2].plot(n, p_lorentzian_0p2, color='orange', label='a=0.2')
ax[1, 2].plot(n, p_lorentzian_0p4, color='yellow', label='a=0.4')
ax[1, 2].plot(n, p_lorentzian_0p6, color='green', label='a=0.6')
ax[1, 2].plot(n, p_lorentzian_0p85, color='blue', label='a=0.85')
ax[1, 2].set_title("Lorentzian (spectroscopy)")
ax[1, 2].legend()

ax[1, 3].plot(n, p_flat_0p2, color='orange', label='a=0.2')
ax[1, 3].plot(n, p_flat_0p4, color='yellow', label='a=0.4')
ax[1, 3].plot(n, p_flat_0p6, color='green', label='a=0.6')
ax[1, 3].plot(n, p_flat_0p85, color='blue', label='a=0.85')
ax[1, 3].set_title("Flat (Uniform)")
ax[1, 3].legend()
```

# <font color='lightblue'> Equilibrium </font>

> Goal

* Nature tries to <font color='red'>minimize energy</font> and <font color='red'>maximize entropy</font>
* To predict these tendencies, we need to to compute extrema or <font color='red'>States of equilibrium</font>

---

> Degrees of Freedom (DoFs)

Any quantity (or variable) that is free to change in the system. From classical Thermo, this is governed by <font color='red'>Gibbs Phase Rule</font>:

$$
F = C - P + 2
$$

where 
* F = Degrees of Freedom 
* C = Number of chemical components 
* P = Number of phases 

<font color='red'>Constraints</font> are the opposites of DoFs. A system governed by an extremum principle will change its DoFs until it reaches the maximum or minimum allowed by the constraint.

--- 

> Equilibrium 

Equilibrium describes the state, $x^*$,  where a system tends to go and stay (i.e. become stationary) since the <font color='red'>net force is 0</font>. While the magnitude of the force describes the *strength of it's tendency* towards equilibrium, stability also depends on the response to perturbations.


* <font color='red'>Stable equilibrium state</font>: Potential energy, V(x), is at a *global minimum*
    * Always decreasing: $V(x) > V(x^*)$,  $\forall x \neq x^*$
    * Global minimum: $\frac{dV(x)}{dx} = 0$ @ $x = x^*$
    * Curves up [convex]: $\frac{d^2V(x)}{dx^2} >0$

* <font color='red'>Neutral</font> equilibrium State: Potential energy, V(x), is *flat*
    * Flat: $\frac{dV(x)}{dx} = 0$ $\forall x$.

* <font color='red'>Metastable</font> equilibrium State: Potential energy is at a *local minimum*
    * Local stability: $V(x) > V(x^*)$ for small $\epsilon = |x-x^*|$
    * Local minimum: $\frac{dV(x)}{dx} = 0$ @ $x = x^*$
    * Local stability: $V(x) < V(x^*)$ for large $\epsilon = |x-x^*|$

* <font color='red'>Unstable</font> equilibrium State: Potential energy is at a *maximum*
    * Mountain peak: $V(x) < V(x^*)$, $\forall x \neq x^*$
    * Global maximum: $\frac{dV(x)}{dx} = 0$ @ $x = x^*$
    * Curves down [concave]: $\frac{d^2V(x)}{dx^2}<0$




<font color='purple'>Note: This is different from Steady state, where things can still move around</font>

--- 

> Extremum Principle

<font color='red'>Maximizing multiplicity</font> predicts the most probable outcome [i.e. Law of large numbers].

---

> Equilibrium Forces / Driving Potentials

* Pressure: Mechanical Equilibrium $\implies$ Force for expansion (related to [PV] work)
* Chemical Potential: Chemical Equilibrium $\implies$ Force for diffusion / rxns (related to Gibbs free energy)
* Temperature: Thermal Equilibrium $\implies$ Force for heat transfer (related )

# <font color='lightblue'> Laws of Thermodynamics</font>

> Goal 

<font color='red'>Energy</font> [units: joules (J), calories (cal), Watt-hours (Wh), British Thermal Units (BTUs)] and <font color='red'>Power</font> [energy per time, Watts (W) = $\frac{J}{s}$] are important metrics to understand commerce, technology, the environment, climate, life and more. WHen we say "energy consumption", we mean that it has *transferred* from a captured from to a useable one. 

* In the 1800s, the main question was: "<font color='red'>Why does *heat* flow from hot objects to colder ones</font>"? 
* The accepted answer was that heat was a conserved fluid, until classical Thermo was developed.
* 

<font color='purple'>Note: The ability to *flow* is an aspect of all conserved properties (i.e. mass, momentum, and energy).</font> Conserved $\implies$ flow, but flow does not $\implies$ conserved (e.g. heat)

---

> Law of conservation of Energy

Total energy is the conservation of 
* Kinetic Energy (K): The amount of work an object can do based on it's *motion*
* Potential Energy (V): The amount of work an object can do based on it's *spatial position*

$$
E_{tot} = \frac{mv^2}{2} + V
$$

* We can also replace this with momentum, $p=mv$ to get: 

$$
E_{tot} = \frac{p^2}{2m} + V
$$


---

> Important Experiments

These are experiments that changed the notion that heat is a fluid
* J Black (1760) <font color='red'>Heat Capacity</font>: Heating mercury and water over the same flame resulted in different temperatures, so different materials had a different capacity to hold heat. 
* J Black (1760) <font color='red'>Latent Heat</font>: Water can be heated at 100C, and instead of getting hotter it changes phase at the same temperature, so heat and temperature must be different.
* Count Rumford (1798) <font color='red'>Cannons</font>: Mechanical work from boring cannons was converted to heat.
* H Davy (1799) <font color='red'>Melting Ice</font>: Mechanical work of rubbing ice was enough to melt them.
* Sir William Hersche / John Tyndall (1800) <font color='red'>Radiant Heat</font>: Electromagnetic waves cause heat to move through a vacuum, but vacuums are void of matter, so heat can not be a form of matter
* J Ericsson (1853) <font color='red'>Calorique Theory Piston</font>: 'Regenerator' coils were supposed to collect exhaust heat and reuse it, but second law says perpetual motion is impossible, so this engine failed.
* JP Joule (1850s) <font color='red'>Series of experiments</font>: Multiple experiments showing that heat could be produced from work that made it impossible to ignore any further.

--- 

> 1st Law: Conservation of energy

Energy (and related quantities) describe a systems ability to perform work. Came from a German physicist [Julius Robert (JP) von Mayer] in 1842 while studying oxidation in blood, but was published by the more famous [Rudolf Classius](https://sites.pitt.edu/~jdnorton/teaching/2559_Therm_Stat_Mech/docs/Clauius%20The_Mechanical_Theory_of_Heat%201867.pdf) and William Thomson. It relates the <font color='red'>change in internal energy</font>, $\Delta U$, to measurements of <font color='red'>Heat</font>, Q, and <font color='red'>work out</font>, W.


$$
\Delta U = Q + W_{by} = (Q_{in} - Q_{in}) + (W_{in} - W_{out})
$$

<font color='purple'>Note: Energy is a property of the *system*, while the others are properties of *transferred across boundaries*</font>

---

> Kintetic Theory of Gases / Quantum

Around the late 1800s, scientists (Clausius, Maxwell, Boltzmann) began asking the question: "What is the microscopic nature of heat"? They introduced 3 novel ideas:

1. Matter is made of freely moving particles in empty space (not the ether)
2. Heat is the transfer of energy from *molecule collisions*
3. Electromagnetism can influence motion (radiant heat)

This was extremely successful in predicting the ideal gas law, diffusion rates, viscosities, thermal conductivities, etc. 

$$
\frac{3}{2}kT = \frac{m \langle v^2 \rangle}{2}
$$


This was eventually replaced by the theory that internal energy, U, is quantized as a result of all particles, N, at specific translations, rotations, vibrations, and electronic excitations energy levels, $\epsilon_i$.

$$
U = \sum_{i}N_i \epsilon_i
$$



---

> 2nd Law: Tendency towards Equilibrium

However, energy balances don't explain <font color='red'>Why heat flows</font>. Instead, this tendency is a result of nature to *maximize multiplicity*, which is explained by the $2^{nd}$ Law of Thermo

# <font color='lightblue'> Calculus</font>

> Goal 

We have solved problems with 1 DoF, now let's review tools from multivariate calculus to help us with multiple DoFs. The two main tools from calculus are differentiation (find extrema) and integration (find areas).

--- 

> Partial Derivatives

A function of 2+ variables represents a *surface* (as opposed to a curve) and the partial derivative is the <font color='red'>slope along *one direction*</font>.

$$
(\frac{\partial f(x,y)}{ \partial x})_y = lim_{\Delta x \to 0} \frac{f(x+\Delta x, y) - f (x,y)}{\Delta x}
$$

---

> Total Differential

Recall that the Taylor Series Expansion, centered at a point a, approximates a function as: 

$$
f(x) = \sum_{i=0}^n \frac{f^i(a)(x-a)^i}{i!} = f(a) + f'(a)\frac{(x-a)}{1!} + f''(a)\frac{(x-a)^2}{2!} + ... + f^{n}(a)\frac{(x-a)^n}{n!}
$$

We can use this to show that

$$
\Delta f = f(x) - f(a) = f'(x) \Delta x + f''(x) \frac{\Delta x^2}{2} + f^{(3)}(x) \frac{\Delta x^3}{6} + ...
$$

And taking the $lim_{\Delta x \to 0}$, we get: 

$$
df = f'(x) dx
$$

This trend holds in bivariate functions, giving us the total derivative: 

$$
df = \frac{\partial f}{\partial x}|_y dx + \frac{\partial f}{\partial y}|_x dy 
$$

In general, for multivariate functions, this is: 

$$
df = \sum_{i=1}^t \frac{\partial f}{\partial x_i}|_{j \neq i} dx_i 
$$

<font color='purple'>Note: This relates EXACT differentials, with PARTIAL differentials</font>

---

> Extrema of Multivariate Functions 

In the univariate case, we have 1 variable and 1 derivative to be equal to zero. For multivariate cases, we have t variables and the extrema is only satisfied if we simultaneously set t partial derivatives equal to 0. 

$$
\frac{\partial f}{\partial x_i}|_{j \neq i} = 0, \forall i=1,2,...,t
$$

---

> Lagrange Multipliers

* <font color='red'>Constraint equations</font>, $g(x_1, x_2, ...., x_n) = c$, force a relationship between multiple DoFs to *remove independency* (ex: Ideal gas law forces V and T to be inversely proportional, while Peng-Robinson has a stronger constraint)

* We must satisfy 2 conditions simulatneously: 
    * Total Differential of <font color='red'>original function</font> is 0: $df(x,y) = 0$
    * Total Differential of <font color='red'>constraint function</font> is 0: $dg(x,y) = 0$
    * <font color='green'>Algebra left as excercise for reader, but see pg 69 in Dill and Bromberg</font>

* We call the <font color='red'>ratio of derivatives</font> an arbitrary constant known as the *Lagrange Multiplier*

* In general, each new constraint equation adds a Lagrange Multiplies: 

$$
df = \sum_{i=1}^t \frac{\partial f}{\partial x_i} d x_i = 0
$$

$$
dg = \sum_{i=1}^t \frac{\partial g}{\partial x_i} d x_i = 0
$$

$$
dh = \sum_{i=1}^t \frac{\partial h}{\partial x_i} d x_i = 0
$$

$$
d(f - \alpha g - \beta h) = \sum_{i=1}^t [\frac{\partial f}{\partial x_i} - \alpha \frac{\partial g}{\partial x_i} - \beta \frac{\partial h}{\partial x_i} ]d x_i = 0
$$


* This can be rewritten in vector notation simply as: 

$$
g(x_1, x_2, ..., x_n) = c
$$

$$
\nabla f - \lambda \nabla g = 0
$$



---

> Pathways of Integration 

* Pathway: A particular sequence of values for the independent variables when you integrate them. For a 2 variable equation, it has the general form:

$$
\text{path} = \int_{y_a}^{y_b} \int_{x_a}^{x_b} [ t(x,y) dx + s(x,y)dy]
$$


* <font color='red'>State</font> function: Regardless of how you integrate, the values will be the same [i.e has an <font color='red'>exact differential</font>] (e.g. Altitude)

$$
\int_{y_a}^{y_b} \int_{x_a}^{x_b} [ t(x,y) dx + s(x,y)dy] = \int_{x_a}^{x_b} \int_{y_a}^{y_b} [ t(x,y) dx + s(x,y)dy]
$$

* Example: $\text{path} = ydx + xdy$

* <font color='red'>Path-dependent</font> function: Order of integration leads to unique values (e.g. total distance)

$$
\int_{y_a}^{y_b} \int_{x_a}^{x_b} [ t(x,y) dx + s(x,y)dy] \neq \int_{x_a}^{x_b} \int_{y_a}^{y_b} [ t(x,y) dx + s(x,y)dy]
$$

* Example: $\text{path} = xy^2dx + xdy$

* Euler's Reciprocal Test: From Differential Equations, we know we can verify if an equation is an exact differential by using the Schwarz's (Clairaut's) Theorem:

$$
\frac{\partial^2 f}{\partial x \partial y} = \frac{\partial^2 f}{\partial y \partial x}
$$

<font color='purple'>State functions need t(x) only and s(y) only, so that t(x)dx and s(y)dy makes an exact differential </font>

---

> Chain Rule

This is used to get the derivatives of *composite functions*. We start with the <font color='red'>total derivative</font>:

$$
df = \frac{\partial f}{\partial x}_{y,z} dx + \frac{\partial f}{\partial y}_{x,z} dy + \frac{\partial f}{\partial z}_{x, y} dz
$$

Then consider the differentials with respect to a *latent* variable, u:

$$
dx = \frac{dx}{du}du, dy = \frac{dy}{du}du, dx = \frac{dz}{du}du
$$

Substitute:
$$
df = \frac{\partial f}{\partial x}_{y,z} \frac{dx}{du}du + \frac{\partial f}{\partial y}_{x,z} \frac{dy}{du}du + \frac{\partial f}{\partial z}_{x, y} \frac{dz}{du}du
$$

Simplify: 
$$
\frac{df}{du} = \frac{\partial f}{\partial x}_{y,z} \frac{dx}{du} + \frac{\partial f}{\partial y}_{x,z} \frac{dy}{du} + \frac{\partial f}{\partial z}_{x, y} \frac{dz}{du}
$$

# <font color='lightblue'> Summary</font>

> Probability

* Probability: $\mathbb{P}_A = \frac{n_A}{N} = \frac{\text{number of outcomes that falls into category A}}{\text{total number of outcomes}}$
    * *Deterministic*: Only 1 outcome is possible [$\mathbb{P}_A =1$]
    * *Mutually Exclusive*: A and B can never happen together [$\mathbb{P}(A \cap B) = 0$]
    * *Collectively Exhaustive*: A set that includes all possible outcomes, leaving no gaps [$S = \cup_{i=1}^n A_i$]
    * *Independent*: Outcomes are unrelated (not correlated) [$\mathbb{P}(A | B) = \mathbb{P}(A)$]
    * *Multiplicity*: The total number of ways different outcomes can occur [$W = \prod_{i=events} n_i$]

* Rules: 
    * *Addition Rule*: <font color='red'>A OR B OR ...</font> [$\mathbb{P}_{A \cup B} = \frac{n_A + n_B - n_{A,B}}{N} = \mathbb{P}_A + \mathbb{P}_B - \mathbb{P}_{A \cap B}$] Watch for *MUTUALLY EXCLUSIVE*
    * *Multiplication Rule*: <font color='red'>A AND B AND ...</font> [$\mathbb{P}_{A \cap B} = \mathbb{P}_A \mathbb{P}_{B|A} = \mathbb{P}_B \mathbb{P}_{A|B}$] Watch for *INDEPENDENT*

* Composite events: A set of elemenetary events (e.g. coin flip, dice roll) collected together [or *multiplication of a bunch of failures*]
    * Independent events but of the or type
        * Ex 1: 1 on first roll or 4 on second roll = 11/36
        * Ex 2: Tails on the first flip or heads on the third flip = 6/8
        * Ex 3: If I have 23 people in a room, what is the probability that 2 will share a birthday = 50% [<font color='green'>Birthday paradox</font>]

* Degree of correlation: How much does A influence B $\implies g = \frac{\mathbb{P}_{B|A}}{\mathbb{P}_B} = \frac{\mathbb{P}_{AB}}{\mathbb{P}_A \mathbb{P}_B}$


--- 

> Combinatorics

* Permutations: The <font color='red'>total numer of ways</font> to arrange a set of items (order invariant) [$W = \frac{N!}{n_1! n_2! ... n_t!} = \frac{N!}{\prod_{i} n_i}$]
* Multiplicity: Number of ways to arrange a set of items <font color='red'>with a specific outcome</font> (i.e. total internal energy) [n items, N positions $\implies W(n,N) = \binom{N}{n}$]
* Normailzation: When PDFs <font color='red'>don't sum to 1</font> [$p(x) = \frac{g(x)}{\int_a^b g(x)dx}$]
* Binomial Distribution: $\text{Bin}(n,N) = \binom{N}{n} \mathbb{P}^n (1-\mathbb{P})^{N-n}$
* Multinomial Distribution: $\text{Multi}(n_1, n_2, ..., n_t, N) = \frac{N!}{n_1! n_2! ... n_t!} \mathbb{P}_1^{n_1} \mathbb{P}_2^{n_2} ... \mathbb{P}_t^{n_t}$

---

> Statistics 

* Moments: <font color='red'>Describes the shape</font> of the distribution $\langle x^n \rangle = \int_a^b x^n p(x)dx$
* Mean (1st Moment): Central tendency 
* Variance (2nd Moment): Spread
* Skewness (3rd Moment): Asymmetry
* Kurtosis (4th Moment): Tail heaviness

* Common Distributions 
    * Bernoulli: $g(n) = \mathbb{P}^n (1-\mathbb{P})^{N-n}$
    * Binomial: $p(n) = \binom{N}{n} \mathbb{P}^n (1-\mathbb{P})^{N-n}$
    * Poisson: $p(n) = \frac{x^n}{n!}e^{-x}$ <font color='purple'>Note: Taylor series expansion</font> $e^x = \sum_n \frac{x^n}{n!}$
    * Gaussian: $p(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x-\mu)^2}{2 \sigma^2}}$
    * Exponential (Boltzmann): $p(x) = ce^{-cx}$ where $0 \leq x \leq \infty$
    * Power Law: $p(x) = \frac{1}{x^c}$
    * Lorentzian: $p(x) = \frac{c}{\pi [(x - \langle x \rangle)^2 + c^2]}$ [Useful in spectroscopy]
    * Flat: $p(x) = \frac{1}{c}$

----

> Equilibrium 

* Degrees of Freedom: Any variable that can change (e.g. Temperature, Pressure, Volumne, Moles, etc.)
* Types of Equilibrium:
    * Stable: Bottom of a <font color='red'>convex energy well</font>
    * Neutral: Flat energy well
    * Metastable: Bottom of energy well, but there can be lower bottoms
    * Unstable: Top of a <font color='red'>concave energy well</font>

* Extrema Principle: Nature is trying to <font color='red'>maximize multiplicity</font>

---

> Laws of Thermo 

* <font color='red'>ALL CONSERVED PROPERTIES FLOW, BUT ALL THINGS THAT FLOW ARE NOT CONSERVED PROPERTIES</font>
* 1st Law: Conservation of Energy [$\Delta U = Q + W_{by}$]
* Kinetic Theory of Gases: Matter is free particles that transfer heat at collision [$\frac{3}{2}kT = \frac{m \langle v^2 \rangle}{2}$]
* 2nd Law: Heat flows to *maximize multiplicity*

---

> Multivariate Calculus

* Total Differential: $df = \sum_{i=1}^t \frac{\partial f}{\partial x_i}|_{j \neq i} dx_i $
* Multivariate Extrema: $\frac{\partial f}{\partial x_i}|_{j \neq i} = 0, \forall i=1,2,...,t$
* Lagrange Multipliers: $\nabla f - \lambda \nabla g = 0$
* Pathways of Integration:
    * <font color='red'>State function</font>: Exact differentials, need Euler recipricol relationship [$df = tdx + sdy = \frac{\partial f}{\partial x}dx + \frac{\partial f}{\partial y}dy$]
    * <font color='red'>Path-dependent function</font>: Inexact differentials [$df = t(x,y)dx + s(x,y)dy$]
* Chain Rule: For composite functions $\frac{df}{du} = \frac{\partial f}{\partial x}_{y,z} \frac{dx}{du} + \frac{\partial f}{\partial y}_{x,z} \frac{dy}{du} + \frac{\partial f}{\partial z}_{x, y} \frac{dz}{du}$
